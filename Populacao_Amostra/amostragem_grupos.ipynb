```python
import pandas as pd
import numpy as np
import random

dataset = pd.read_csv('census.csv')

def amostragem_grupo(dataset, numero_grupos):
  intervalo = len(dataset)/numero_grupos
  grupo = []
  id_grupo = 0
  contagem = 0    
  for _ in dataset.iterrows():
     grupo.append(id_grupo)
     contagem +=1
  if contagem > intervalo:
      contagem = 0
      id_grupo +=1
      dataset['grupo'] = grupo
      random.seed(1)
      grupo_selecionado = random.randint(0, numero_grupos)
      return dataset[dataset['grupo']== grupo_selecionado]

df_amostra_agrupamento = amostragem_grupo(dataset, 10)
print(df_amostra_agrupamento)
print(df_amostra_agrupamento.shape, df_amostra_agrupamento['grupo'].value_counts())
```
