```python


import pandas as pd
import numpy as np
import random

dataset = pd.read_csv('census.csv')

def amostragem_simplificada(dataset, amostra):
  intervalo = len(dataset)/ amostra
  random.seed(1)
  inicio = np.random.randint(0, intervalo)
  indices_amostras = np.arange(inicio, len(dataset), step = intervalo)
  amostra_sistematica = dataset.iloc[indices_amostras]
  return amostra_sistematica

finalizando_amostra = amostragem_simplificada(dataset, 100)
print(finalizando_amostra)

```
