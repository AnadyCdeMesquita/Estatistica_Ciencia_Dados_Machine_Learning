```python
import pandas as pd
import numpy as np
import random
from sklearn.model_selection import StratifiedShuffleSplit

dataset = pd.read_csv('credit_data.csv')
#display(dataset)
print(dataset['age'].mean())
  
#ALEATORIA

def aleatoria(dataset, n):
  return dataset.sample(n)

amostragem_aleatoria = aleatoria(dataset, 1000)
# print(amostragem_aleatoria)

print(amostragem_aleatoria['age'].mean())


#SISTEMATICA

def estratificada(dataset, n):
  intervalo = len(dataset) / n
  inicio = np.random.randint(0, intervalo)
  amostras = np.arange(inicio, len(dataset), step = intervalo)
  amostra_sistematica = dataset.iloc[amostras]
  return amostra_sistematica
  
resultado = estratificada(dataset, 100)
# print(resultado)

print(resultado['age'].mean())


#ESTRATIFICADA
tamanho_amostra_Porcentagem = 1000/len(dataset) # OS 100 SIGNIFICA O TAMANHO QUE SE QUER DA AMOSTRA
print(tamanho_amostra_Porcentagem)
def estratificada(dataset, tamanho_amostra_Porcentagem):
  split = StratifiedShuffleSplit(test_size= tamanho_amostra_Porcentagem) 
  for _ ,  y in split.split(dataset, dataset['c#default']): #O X é o tabelão e Y é 100 amostras
   df_y = dataset.iloc[y]
  return df_y

df_amostra_estratificada = estratificada(dataset,0.05 )
print(df_amostra_estratificada)
# print(df_amostra_estratificada.shape)

print(df_amostra_estratificada['age'].mean())

```
