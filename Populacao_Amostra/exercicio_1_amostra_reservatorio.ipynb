```python
import pandas as pd
import numpy as np
import random


dataset = pd.read_csv('credit_data.csv', chunksize=1)

# RESERVATÓRIO
random.seed(42)

def amostragem_reservatorio(dataset, tamanho_amostra):
    reservatorio = []

    for i, linha in enumerate(dataset):
        # O pandas lê o chunk como um DataFrame de 1 linha, então pegamos o dicionário dela
        dados_linha = linha.iloc[0].to_dict()

        # Passo 1: Se o reservatório ainda não está cheio, colocamos a linha direto lá
        if i < tamanho_amostra:
            reservatorio.append(dados_linha)
        else:
            # Passo 2: Para as próximas linhas, sorteamos um índice de 0 até i
            j = random.randint(0, i)

            # Passo 3: Se o índice sorteado for menor que o tamanho do reservatório,
            # o elemento antigo daquela posição é substituído pelo novo!
            if j < tamanho_amostra:
                reservatorio[j] = dados_linha

    # No final, transformamos nossa lista de dicionários de volta em um DataFrame bonitinho
    return pd.DataFrame(reservatorio)


# --- Executando o nosso teste ---
# Vamos puxar uma amostra de 10 linhas da tabela 'census.csv'

amostra_final = amostragem_reservatorio(
    dataset, 1000
)


# Mostrando o resultado
print(f"Sucesso! Criamos uma amostra com {len(amostra_final)} linhas.")
print("\nAqui está um pedaço da nossa amostra:")
print(amostra_final)


print(amostra_final['age'].mean())

```
