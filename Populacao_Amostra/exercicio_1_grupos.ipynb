```python
import pandas as pd
import numpy as np
import random

df = pd.read_csv('credit_data.csv')

random.seed(42)
def amostragem_grupos(df, n_grupos):
  
    # Calcula o tamanho exato de cada grupo
    tamanho_grupo = len(df) / n_grupos

    # Cria os IDs dos grupos de forma vetorizada e rápida
    grupos = (np.arange(len(df)) / tamanho_grupo).astype(int)

    # Garante que nenhum índice passe do limite máximo de grupos (0 a n_grupos-1)
    grupos = np.clip(grupos, 0, n_grupos - 1)

    # Atribui a nova coluna ao DataFrame
    df_resultado = df.copy()
    df_resultado["grupo"] = grupos
  
    return df_resultado

dataset_com_grupos = amostragem_grupos(df, 2)

print(dataset_com_grupos)

print(dataset_com_grupos['grupo'].value_counts())
print(dataset_com_grupos.shape)

print(dataset_com_grupos['age'].mean())
```
